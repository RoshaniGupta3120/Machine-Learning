### Null removal steps

```python
# replacing ? with NaN (Not a number)
df["normalized-losses"]=df["normalized-losses"].replace("?", np.nan)
df["horsepower"]=df["horsepower"].replace("?", np.nan)
df
```

```python
# np.nan data type is float that's why we typecasted the columns to float
df["normalized-losses"]=df["normalized-losses"].astype("float")
df["horsepower"]=df["horsepower"].astype("float")
df.info()
```

```python
df.isnull().sum()  # total numbers of missing values
(df.isnull().sum()/len(df))*100  # -- % of missing values in each column
'''
if % of missing values is between 0.1-3 ---- drop those rows
4%-40% ---- for numerical columns -- if normal distribution, replace with mean()
                                     if skweness is there , replace with median()
            for categorical columns --- replace with mode()
<40% ----- drop that column
'''
```

```python
df[["normalized-losses","horsepower"]].describe()
# if mean > median ----- positve skewed
# if mean < median ----- negative skewed

# horsepower --- mean > median ----- positve skewed
# normalized-losses --- mean > median ----- positve skewed
```

```python
plt.figure(figsize=(2,2))
df["normalized-losses"].plot(kind='kde')
plt.show()
# the chart shows a +ive skewness in normalized-losses col
```

### Skewness Removal

# Rules for removing Skewness
1. if the column which has skewness is showing good co-relation with the target then do not remove the skewness
2. If the column which has skewness has negative values then do not remove the skewness

```python
df[c].corr()["price"].sort_values()
'''
highly negative correlated --
highway-mpg         -0.704846
city-mpg            -0.680412
highly positve correlated --
width                0.730503
horsepower           0.769844
engine-size          0.869638

According to rulesss--------------------
we'll only remove skewness from
normalized-losses -- 0.9632059441735282
'''
```

```python
# removing skewness from "normalized-losses" by using log
df["normalized-losses"]=np.log(df["normalized-losses"])
print(skew(df["normalized-losses"]))
df["normalized-losses"].plot(kind="kde")
plt.show()
```

```python
features=df.drop('price', axis=1)
target=df['price']
```

```python
print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,random_state=10,test_size=0.2)
print(f"Training data : {xtrain.shape,ytrain.shape}")
print(f"Testing data : {xtest.shape,ytest.shape}")
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(ytest,ypred)
r2=r2_score(ytest,ypred)
print("Mean Square error i.e. error in prediction of each point : ", mse)
print("R2 Score : ",r2)
```

```python
lr.score(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```

training score> testing score -- case of overfittin
testing score > traiing score -- case of underfitting.
Underfitting -- increase number of features , do feature engineering, connect with client fo rmore data.
overfitting-- regularization -- to add external error during training time.
2 tpes of regularization --
L1(Lasso Regularization) : = Residual error(RSS)+ lamda+(sum(abs(all slope))) 
lamda -- 200-1000
L2(Ridge Regularization) : = Residual error(RSS)+ lamda+(sum(abs(all slope)**2))#take square of all the slopes
lamda -- 1-50 #numbers are small as you are taking square of slopes

```python
from sklearn.linear_model import Lasso
#hyperparameter tuning
for i in range(200,1001):
    l1=Lasso(i) #hyperparameter tuning
    l1.fit(xtrain,ytrain)
    score_test=l1.score(xtest,ytest)
    score_train=l1.score(xtrain,ytrain)
    print(f"i : {i} train : {score_train} test : {score_test}")
```

```python
l1=Lasso(200) #hyperparameter tuning
l1.fit(xtrain,ytrain)
score_test=l1.score(xtest,ytest)
score_train=l1.score(xtrain,ytrain)
print(f"train : {score_train} test : {score_test}")
```

```python
dict={"Input":xtrain.columns, "slope after applying Lasso" : l1.coef_}
d=pd.DataFrame(dict)
d
```

```python
from sklearn.linear_model import Ridge
#hyperparameter tuning
for i in range(0,51):
    l2=Ridge(i) #hyperparameter tuning
    l2.fit(xtrain,ytrain)
    score_test=l2.score(xtest,ytest)
    score_train=l2.score(xtrain,ytrain)
    print(f"i : {i} train : {score_train} test : {score_test}")
```

```python
dict={"Input":xtrain.columns, "slope after applying Ridge" : l2.coef_}
d=pd.DataFrame(dict)
d
```

### Cross validation
It is another method to deal with overfitting . It is a technique to assist the performance of a machine learning model by splitting the dataset info multiple folds training the model on some folds and testing it on the remaining fold. This ensures robust evaluation by the model also if the testing score by both Rodge and Lasso are same then use cross validatiion to decide which one of them is the best

```python
from sklearn.model_selection import cross_val_score
l1=Lasso(201)
scores = cross_val_score(l1, features, target, cv=5)
scores
```

```python
np.mean(scores).round(2)
```
